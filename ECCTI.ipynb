```python
import math
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

g=1.4; b=1.2; p=1; r=1
```

```python
g_x = np.linspace(-5, -8.5, 100)
g_y = g**2*gaussian(g_x, -7.5, .25) 
b_x = np.linspace(0, -5, 100)
b_y = b**2*gaussian(b_x, -2.5, .25) 
p_x = np.linspace(0, 5, 100)
p_y = p**2*gaussian(p_x, 2.5, .25) 
r_x = np.linspace(5, 8.5, 100)
r_y = r**2*gaussian(r_x, 7.5, .25) 
```

```python
fig, ax = plt.subplots()
ax.plot(g_x, g_y, color = 'lime', label = 'ion #1', linewidth = 3)
ax.plot(b_x, b_y, color = 'deepskyblue', label = 'ion #2', linewidth = 3)
ax.plot(p_x, p_y, color = 'fuchsia', label = 'ion #3', linewidth = 3)
ax.plot(r_x, r_y, color = 'red', label = 'ion #4', linewidth = 3)
ax.set_xlabel(r'Trap Position ($\mu$m)', fontsize = 20)
ax.set_ylabel(r'Stark Shift (kHz)', fontsize = 20)
ax.set_ylim(0)
ax.legend()
fig.tight_layout()
fig.savefig("gauss.svg")
```

```python
fig, ax = plt.subplots()
ax.bar(220, g, color = 'lime', width = .6, label = 'tone #1')
ax.bar(226.5, b, color = 'deepskyblue', width = .6, label = 'tone #2')
ax.bar(233, p, color = 'fuchsia', width = .6, label = 'tone #3')
ax.bar(239.5, r, color = 'red', width = .6, label = 'tone #4')
ax.set_xlabel(r'AOD Drive Tone (MHz)', fontsize = 20)
ax.set_ylabel(r'AOD Drive Power (dB)', fontsize = 20)
ax.legend()
fig.tight_layout()
fig.savefig("frequencies.svg")
```
