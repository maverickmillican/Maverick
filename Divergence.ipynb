```python
def divergence(distances, x_sizes, y_sizes):
    return 2* np.arctan((((x_sizes[-1]+y_sizes[-1])/2)-((x_sizes[0]+y_sizes[0])/2)) / (2*(distances[-1]-distances[0])))
```

```python
divergence(distances1, x_size1, y_size1)
```

```python
divergence(distances0, x_size0, y_size0)
```

```python
divergence(distances2, x_size2, y_size2)
```

```python
def divergencesum(distances, xlist, ylist):
    this = []
    for i in range(len(distances)):
        for j in range(len(distances)):
            if j > i:
                this.append(2* np.arctan((((xlist[j]+ylist[j])/2)-((xlist[i]+ylist[i])/2))/(2*(distances[j]-distances[i]))))
    return this
print(np.average(divergencesum(distances0, x_size0, y_size0)))
print(np.average(divergencesum(distances1, x_size1, y_size1)))
print(np.average(divergencesum(distances2, x_size2, y_size2)))
```

```python
plt.plot(divergencesum(distances0, x_size0, y_size0))
plt.plot(divergencesum(distances1, x_size1, y_size1))
plt.plot(divergencesum(distances2, x_size2, y_size2))
```
